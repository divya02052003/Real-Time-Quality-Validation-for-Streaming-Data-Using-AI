```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import pickle
from imblearn.over_sampling import SMOTE

# Load dataset
df = pd.read_csv("creditcard.csv")
print("Dataset Preview:\n", df.head())

# Remove duplicate records
df = df.drop_duplicates()
print("After Removing Duplicates:\n", df.shape)

# Normalize 'Amount' column using Min-Max scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['Amount'] = scaler.fit_transform(df[['Amount']])

# Class imbalance handling using SMOTE
X = df.drop(columns=['Class'])
y = df['Class']
smote = SMOTE(sampling_strategy=0.5, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled
)

# Train a Random Forest model
clf = RandomForestClassifier(n_estimators=50, max_depth=10, random_state=42)
clf.fit(X_train, y_train)

# Predictions
y_pred = clf.predict(X_test)

# Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# Save the trained model
with open("fraud_detection_model.pkl", "wb") as model_file:
    pickle.dump(clf, model_file)
print("Model saved successfully as 'fraud_detection_model.pkl'.")

# Save the cleaned dataset
df.to_csv("cleaned_creditcard.csv", index=False)
print("Cleaned dataset saved successfully.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix
disp = ConfusionMatrixDisplay(conf_matrix, display_labels=["Non-Fraud", "Fraud"])
disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
```
